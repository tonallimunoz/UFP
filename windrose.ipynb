```python
from windrose import WindroseAxes
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import numpy as np
import pandas as pd
plt.rcParams.update({'font.size': 16})
```

```python
#Abril
    
apr = pd.read_csv("meteoabril.txt", delimiter = "\t")
apr.index = pd.date_range('4/1/2017', periods = 43200, freq ='Min')
del apr['Fecha']


velocidadapr = apr['m/s'].resample('10Min').mean()
direccionapr = apr['deg'].resample('10Min').mean()

#----------------------------------------------------------------------------

#Mayo

may = pd.read_csv("meteomayo.txt", delimiter = "\t")
may.index = pd.date_range('5/1/2017', periods = 43200, freq ='Min')
del may['Fecha']

    
velocidadmay = may['m/s'].resample('10Min').mean()
direccionmay = may['deg'].resample('10Min').mean()

#----------------------------------------------------------------------------

#Junio

jun = pd.read_csv("meteojunio.txt", delimiter = "\t")
jun.index = pd.date_range('6/1/2017', periods = 43200, freq ='Min')
del jun['Fecha']

    
velocidadjun = jun['m/s'].resample('10Min').mean()
direccionjun = jun['deg'].resample('10Min').mean()
```

```python
#22 abril

ws = velocidadapr[3090:3102]
wd = direccionapr[3090:3102]

ax = WindroseAxes.from_ax()
ax.bar(wd, ws, normed=True, opening=0.8, edgecolor='white')
ax.set_legend()
```

```python
#13 junio

ws = velocidadjun[1764:1790]
wd = direccionjun[1764:1790]

ax = WindroseAxes.from_ax()
ax.bar(wd, ws, normed=True, opening=0.8, edgecolor='white')
ax.set_legend()
```

```python
#24 mayo

ws = velocidadmay[3444:3456]
wd = direccionmay[3444:3456]

ax = WindroseAxes.from_ax()
ax.bar(wd, ws, normed=True, opening=0.8, edgecolor='white')
ax.set_legend()
```
