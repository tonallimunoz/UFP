```python
import pandas as pd
from dateutil import parser
%matplotlib inline
import matplotlib.pyplot as plt
import numpy.ma as ma
```

```python
CPC = pd.read_csv("outputcorrecci√≥n.txt", delimiter = "\t")
CPC.index = pd.date_range('4/1/2017', periods = 780770, freq ='S')
CPCtotal = pd.to_numeric(CPC['CPC']);
CPCpromedio = (CPCtotal.groupby(CPCtotal.index.hour).mean())
```

```python
sumafina = pd.read_csv("channelsumafina.txt", delimiter="\t")
sumafina.index = [parser.parse(b) for b in sumafina['start']]     
del sumafina['start']
sumafinapromedio = sumafina.groupby(sumafina.index.hour).mean()
```

```python
fig, ax1 = plt.subplots(figsize=(9,4))

ax1.set_xlabel('Hour')
ax1.set_ylabel('Particle concentration (p/cm¬≥)')

lns1 = ax1.plot(CPCpromedio, color = "black", label = "CPC")
lns2 = ax1.plot(sumafinapromedio, color = "black", linestyle = 'dashdot', label = "UFPM")

lns = lns1+lns2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=1)
```

```python
sumafinapromedio = pd.to_numeric(sumafinapromedio['SUMA']);
```

```python
ma.corrcoef(ma.masked_invalid(CPCpromedio), ma.masked_invalid(sumafinapromedio))
```
